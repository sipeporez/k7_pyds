## Matplotlib 복습

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 8* np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
a = np.linspace(0, np.pi, 200)
b = np.linspace(0, np.random.randn(), 200)
fig, ax = plt.subplots()
ax.plot(a,b**2);
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
x = np.linspace(0,10,100)
y = np.linspace(0,10,100)
for i in range(10):
    plt.plot(x, np.sin(x-i), 'd')
plt.axis('tight')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.4, cmap='viridis')
plt.colorbar()
```

```python
mean=[0,0]
cov=[[1,1],[1,100]]
x,y=np.random.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x, y, bins=100, cmap='viridis')
cb= plt.colorbar()
```

```python
data = pd.read_csv('data/marathon-data.csv')
print(data.tail(), data.dtypes, sep='\n')
#37249개의 데이터, 시간이 object 타입이기 때문에 변환 필요
#datetime의 timedelta로 데이터 변환, read_csv에서 converters 속성을 통해 데이터를 읽어옴과 동시에 변환 가능
import datetime
def convert_time_type(st):
    h, m, s = map(int, st.split(':'))
    return datetime.timedelta(hours=h, minutes=m, seconds=s)

data = pd.read_csv('data/marathon-data.csv', converters={'split':convert_time_type, 'final':convert_time_type})
print(data.tail(), data.dtypes, sep='\n')

data['split_sec'] = data.split.dt.total_seconds()
data['final_sec'] = data.final.dt.total_seconds()

data.head()
with sns.axes_style('white'):
    g = sns.jointplot(x='split_sec', y='final_sec', data=data, kind='reg', height=10)
    g.ax_joint.plot(np.linspace(4000,16000), np.linspace(8000,32000), ':k') # 검정색 점 모양의 x=y 직선 (동일한 속도로 달린다는 기준)
# 검정선보다 위 => 모든 사람들은 동일한 속도로 달리지 못하고 점점 느려짐 (positive_split)
# kind='reg' => jointplot에서 데이터 예상치를 직접 계산하여 그려주는 속성
```

```python
# 경기 후반부에 더 빨리 달리는 방식 -> negative-split
# split_frac가 0보다 작은 경우 -> negative-split
data['split_frac'] = 1 - (2 * data['split_sec'] / data['final_sec'])
data.head()
```

```python
sum(data.split_frac<0)
# 251명만이 네거티브 스플릿으로 달림
```

```python
sns.displot(data['split_frac'],kde=False, aspect=2);
plt.axvline(0, color='k', linestyle="--"); # 0보다 작은 경우 -> 네거티브 스플릿
plt.axis('tight')
```

```python
# 모든 상관관계에 대한 플롯을 그려주는 PairGrid
g = sns.PairGrid(data, vars=['age','split_sec', 'final_sec', 'split_frac'], hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
# 남성과 여성의 완주시간 분포 비교
sns.violinplot(x="gender", y="split_frac", data=data, hue="gender", palette=["lightblue","lightpink"]);
```

```python
#연령대별 분포 비교
data['age_dec']= data.age.map(lambda age: 10* (age // 10))
data.head()

men = (data.gender == 'M')
women = (data.gender == 'W')
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
with sns.axes_style(style=None):
    sns.violinplot(x='age_dec', y='split_frac', 
                   hue='gender', 
                   data=data, split=True, 
                   inner='quartile',
                   palette=["lightblue","lightpink"])

plt.axhline(0, color='k', linestyle="--");
# 80대의 데이터는 분포수가 적으므로 이상값이 나옴
sum(data.age>=80)
```

### subplot
+ seaborn을 사용하는 것이 더 편함

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4) # 서브플롯간의 간격 조절
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5, 0.5, str((2, 3, i)),fontsize=18, ha='center')
```

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row') # 서브플롯 간의 x축, y축 공유|
```

```python
grid = plt.GridSpec(2, 3) # 2개 이상의 행 또는 열로 확장
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```
